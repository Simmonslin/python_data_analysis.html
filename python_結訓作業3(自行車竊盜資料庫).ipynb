```python
import pandas as pd

datas=pd.read_csv("臺北市10401_11002自行車竊盜點位資訊.csv",encoding="cp950")


real_datas=pd.DataFrame(datas)

print(real_datas)


print()






def data_HeadOrTail(n):
    print()
    
    print("功能: 查詢前數筆資料 or 後數筆資料")
    ans=input("head or tail : ")
    if ans=="head":
        print("查詢前"+str(n)+"筆資料 :")
        print(real_datas.head(n))
        
    elif ans=="tail":
        print("查詢末"+str(n)+"筆資料 :")
        print(real_datas.tail(n))
        


def searcher_region():
    
    print()
    
    print("功能 : 查詢所搜尋區內竊盜案件的地址，並計算案件數量 ")

    place=real_datas["發生(現)地點"]
    keyword=input("請輸入區名 : ")
    
    number=0
    
    
    for i in place:
       
        if keyword in i:
            print(keyword+"內發生過自行車竊盜案件的所有地址: ")
            print(i)
            print()
            number+=1
            
    
    if number==0:
        print("查無此區資料")
    
        
    ans2=input("是否需要%s區內案件發生總次數 ?(請按 Y/N): "%keyword)
    if ans2=="Y":
        print(keyword,"案件總次數 : %d 次" %number)
    else:
        print("資料不顯示")
        


# 很重要，DataFrame中的values能夠以二維串列形式取出資料
# 語法 : .values[] 

print(real_datas.values)

def year_to_month_searcher():
    
    print()
    
    print("功能 : 將搜尋範圍縮小，並整理出特定數據 ")

    keyword=input("請輸入年份 : ")
    
    yes_no=input("是否要將範圍縮小至月份 (Y/N) : ")
    
    if yes_no=="Y" or "y":
        month=input("請輸入月份 : ")
    else:
        month=" "
        
    c=0
        
    for i in real_datas.values:
        a=i[:][2]
        str_a=str(a)
        
        
        if keyword in str_a[0:3]:
            
            if month==" ":
                print("\n發生日期 : %s 年 %s 月 %s 日" %(str_a[0:3],str_a[3:5],str_a[5:7]))
                print()
            
                print("發生時段 : %s" %i[:][3])
                print()
            
                print("發生地點 : %s" %i[:][4])
                print()
                c+=1
            else:
                if keyword in str_a[0:3] and month in str_a[3:5]:
                    print("\n發生日期 : %s 年 %s 日 " %(str_a[3:5],str_a[5:7]))
                    print()
                
                    print("發生時段 : %s" %i[:][3])
                    print()
                
                    print("發生地點 : %s" %i[:][4])
                    print()
                    c+=1
                
    if month==" ":
        print("%s 年共發生 %d 次案件" %(keyword,c))
        
    else:
        print("%s 年 %s 月共發生 %d 次案件" %(keyword,month,c))
            
        


data_HeadOrTail(int(input("請輸入要查詢的數量: ")))

print()

searcher_region()

print()

year_to_month_searcher()

print()


    
```
