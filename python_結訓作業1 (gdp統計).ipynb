```python
import pandas as pd


data_1=pd.read_csv("gdp.csv",header=0)


print(data_1)



ans1=data_1.groupby("country")['country'].count()

print(ans1.sort_values(ascending=False)) 


print()


class gdp:
     def function_continent():
         
        print()
         
        print("功能 : 查詢所輸入的國家是在哪一洲 ")
    
        
        while True:
            
            country=input("Please type the name of the country : ")
            
            if country=="finish":
                break
    
            else:
                dict1=dict()
    
                set1=set()
    
    
                input_country=data_1[data_1.country==country]
        
                dict1.update(input_country["continent"])
    
    
                set1.update(dict1.values())
    
    
                continent=list(set1)
    
                print("%s is in %s" %(country,continent[0]))
            
            
            
            
    
            
            
    
            
    
    
    
    
     def function_year():
         
        print()
         
        print("功能 : 搜尋特定年份的特定國家資料 ")
         
        
        while True:
            
            
            
            year=int(input("Please type a searching year : "))
            country=input("Please type the country : ")
            
            
            data_2=data_1[data_1.country==country]
           
            data_2_year=data_2[data_2.year==year]
            
            if data_2 is None:
               
                print("查無資料")
                
            elif country=="finish":
                break
                
            
            else:
                print(data_2_year)
            

    
    
    
    
     def function_continent_year():
         
        print()
        
        print("功能 : 搜尋特定洲的特定年份資料 ")
        
        while True:
            continent=input("Please choose a continent : ")
            year=int(input("Please choose an existed year : "))
        
            data_c1=data_1[data_1.continent==continent]
            data_c2=data_c1[data_1.year==year]
            
        
            
            if data_c2 is None:
                continue 
            
            elif continent=="finish":
                break
            
            else:
                print(data_c2)
                
    
    
        

   
     def lifeExp():
         print()
         
         print("功能 : 搜尋各國的預期壽命和相關的統計數據")
         
        
         country=input("Please choose a country : ")
         year=int(input("Please select a year : "))
         
        
             
         if year is not None:
             ans1=data_1[data_1.country==country]
             
             
             #很重要 ，若只需要取值，不需要 index 編號，直接加上.values 且不可打 .values()
             
             if year in ans1[ans1.year==year].values:
             
                 ans1_year=ans1[ans1.year==year]
             
                 ans1_lifeExp=ans1_year["lifeExp"]
                 
                 print(ans1_lifeExp.sort_values(ascending=False))
                 
                 print()
                 
                 print("所選取條件下的基本統計數據如下 : ")
                 
              # .describe() 功能 --> 包括 std, count, mean, 25%, 50%, 75%, max, min 
              
                 print(ans1_lifeExp.describe())
                
             
             else:
                 
                 ans1_all=ans1.sort_values(by="lifeExp",ascending=False)
                 
                 print(ans1_all)
                 
                 print()
                 
                 print("詳細資料如下 : ")
                 
                 print(ans1_all["lifeExp"].describe())
                
                
             
             
             
     function_continent()
     
     function_year()
     
     function_continent_year()
        
        
     lifeExp()
        
```
