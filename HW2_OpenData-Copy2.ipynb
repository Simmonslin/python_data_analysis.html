```python
import xml.etree.ElementTree as ET 



class defqon_1:
    def __init__(self):
        self.list1=[]
        self.list2=[]
        self.list3=[]
        self.list4=[]
        self.column=["row_number","cat2","stitle","mrt"]
        
       
        
    def filename(self,file_name):
        datas=ET.parse(file_name)
        root=datas.getroot()
        
        for i in root:
            row_number=i.find("RowNumber").text 
            cat2=i.find("CAT2").text 
            stitle=i.find("stitle").text 
            mrt=i.find("MRT")
            
            if mrt is not None:
                mrt1=mrt.text 
                
            self.list1.append(row_number)
            self.list2.append(cat2)
            self.list3.append(stitle)
            self.list4.append(mrt1)
            self.data={'row_number':self.list1,
                   'cat2':self.list2,
                   'sttile':self.list3,
                   'mrt':self.list4}
            
    def show(self):
        for j in range(len(self.list1)):
            print(self.list1[j]," ",self.list2[j]," ",self.list3[j]," ",self.list4[j]) 
            
      
    def display_column(self):
        for i in range(len(self.column)):
            print(self.column[i],end="  ")
            
    def sum_rownumber(self):
        total=0
        for i in range(len(self.list1)):

            total+=int(self.list1[i])
        
        print("Total of the row numbers: %d" %total)
        
        
        
    def filter_1(self,name,keyword):
        column=self.data[name]
        result=[]
        n=len(self.list1)
        for i in range(n):
            num=self.list1[i]
            yes_no=keyword in column[i]   #yes_no 為 bool 值
            print(num," ",column[i]," ",yes_no)
            result.append(yes_no)
        return result 
    
    
   
     
    #很重要，yes_no=keyword in columns[i] 的 [i] 一定要加，否則 yes_no 的 bool 值皆為 True 
    
    def searcher(self,name,keyword):
        columns=self.data[name]
        for i in range(len(self.list1)):
            yes_no=keyword in columns[i]
            if yes_no==True:
                print(self.list1[i]," ",self.list2[i]," ",self.list3[i]," ",self.list4[i])
```
